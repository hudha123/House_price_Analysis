#### Mean Function

```python
print("price_per_sqft_mean :" ,df.price_per_sqft.mean())
print("-----------------------------")
print("price_per_sqft_median :" ,df.price_per_sqft.median())
print("-----------------------------")
print("price_per_sqft_mode :" ,df.price_per_sqft.mode())
print("-----------------------------")
print("price_per_sqft_variance :",df.price_per_sqft.var())
print("-----------------------------")
print("price_per_sqft_std :" ,df.price_per_sqft.std())
```

Here, mean>median>mode (ie it is positively skewed distribution) 

#### 2)Percentile Method

```python
df.price_per_sqft.quantile([0.25,0.50,0.75])
```

```python
df.price_per_sqft.quantile([0.1,0.2,0.8,0.85,0.87,0.9,0.91,0.95])
```

```python
df_no_outliers1=df[df.price_per_sqft<=df.price_per_sqft.quantile(0.91)]
df_no_outliers1
```

#### IQR(Inter quartile range method

```python
q1=df.price_per_sqft.quantile(0.25)
q3=df.price_per_sqft.quantile(0.75)
print(q1,q3)
IQR=q3-q1
print(IQR)
```

```python
lower_whisker=q1-(1.5*IQR)
upper_whisker=q3+(1.5*IQR)
print(lower_whisker,upper_whisker)
```

```python
df_no_outliers2=df[(df.price_per_sqft>=lower_whisker) & (df.price_per_sqft<=upper_whisker)]
df_no_outliers2
```

#### Normal Distribution

```python
m=df.price_per_sqft.mean()
s=df.price_per_sqft.std()
print(m,"   ",s)
print((m-(3*s)),"   ",(m+(3*s)))
```

```python
df_no_outliers3=df[(df.price_per_sqft>=-325566.06084694766 ) & (df.price_per_sqft<=341831.3445273032)]
df_no_outliers3
```

#### Zscore Method

```python
df["zscore"]=(df.price_per_sqft-m)/s
df["zscore"]
df_no_outliers4=df
df_no_outliers4
```

```python
from scipy import stats
z_scores = stats.zscore(df.price_per_sqft)    # Calculate Z-scores
df_no_outliers5 = df[(z_scores >= -3) & (z_scores <= 3)]   # Identify and remove outliers using Z-scores
df_no_outliers5
```
